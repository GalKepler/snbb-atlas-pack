```python
import nibabel as nib
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
full_hcpex_df = pd.read_csv("/media/storage/Projects/snbb-atlas-pack/atlases/atlas-HCPex/atlas-HCPex_dseg.tsv", sep="\t")
full_mmp_df = pd.read_csv("/media/storage/Projects/snbb-atlas-pack/atlases/atlas-HCPMMP/atlas-HCPMMP_dseg.tsv", sep="\t")
full_nifti_path = Path("/media/storage/Projects/snbb-atlas-pack/atlases/atlas-HCPex/atlas-HCPex_space-MNI152NLin2009cAsym_res-01_dseg.nii.gz")

hcpex_ext_df = full_hcpex_df[full_hcpex_df["mmp_region"].isna()]
```

```python
# generate a new subcortical atlas
full_img = nib.load(full_nifti_path)
full_data = np.round(full_img.get_fdata()).astype(int)

subcortex_labels = hcpex_ext_df["index"].values
is_subcortex = np.isin(full_data, subcortex_labels)
subcortex_data = np.zeros_like(full_data)
subcortex_data[is_subcortex] = full_data[is_subcortex]
subcortex_img = nib.Nifti1Image(subcortex_data, full_img.affine, full_img.header)

subcort_atlas_name = "atlas-HCPexSubcortex"
subcort_atlas_dir = Path("/media/storage/Projects/snbb-atlas-pack/sourcedata") / subcort_atlas_name
subcort_atlas_dir.mkdir(exist_ok=True)
subcort_atlas_path = subcort_atlas_dir / f"{subcort_atlas_name}_space-MNI152NLin2009cAsym_res-01_dseg.nii.gz"
nib.save(subcortex_img, subcort_atlas_path)
subcort_df = hcpex_ext_df.copy()
subcort_df["index"] = subcort_df["index"].astype(int)
subcort_df.to_csv(subcort_atlas_dir / f"{subcort_atlas_name}_dseg.tsv", sep="\t", index=False)
```
